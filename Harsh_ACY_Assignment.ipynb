```python
data = df
```

```python
sns.countplot(x='Class', data=data)
plt.title('Class Distribution Before SMOTE')
plt.show()
```

```python
X = data.drop(['Class'], axis=1)
y = data['Class']
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Visualize class distribution after SMOTE
sns.countplot(x=y_train_smote)
plt.title('Class Distribution After SMOTE')
plt.show()
```

```python
# Function to evaluate models
def evaluate_model(y_true, y_pred, model_name):
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_true, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()
```

```python
# Random Forest Model
rf = RandomForestClassifier(n_estimators = 2, max_depth = 10, random_state=42)
rf.fit(X_train_smote, y_train_smote)
rf_pred = rf.predict(X_test)
evaluate_model(y_test, rf_pred, 'Random Forest')
```

```python
# XGBoost Model
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train_smote, y_train_smote)
xgb_pred = xgb.predict(X_test)
evaluate_model(y_test, xgb_pred, 'XGBoost')
```

```python
# ROC Curve and AUC Comparison
models = {'Random Forest': (rf, rf_pred), 'XGBoost': (xgb, xgb_pred)}
plt.figure(figsize=(10, 7))
for model_name, (model, pred) in models.items():
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
    auc_score = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# Summary of Results
print("\nSummary of Model Performances:")
for model_name, (model, pred) in models.items():
    auc_score = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    print(f"{model_name}: AUC = {auc_score:.2f}")
```
